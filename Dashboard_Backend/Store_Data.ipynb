```python
#Enter the subfolder to the evaluated runs (keep the r in front of the string)
path='02_Dashboard'

#Select ending of the ..._analysis_part1.xls file i.e. 'xls' or 'xlsx'
ending='xlsx'

#Enter the path to the Run information file (keep the r in front of the string)
info_path=r'C:\Users\Jeremias Gerner\Documents\Dash\Incidentlist TVKLVM Simulations.xlsx'



import os
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
import numpy as np
import pandas as pd 
from pandas import DataFrame
import numpy as np
import openpyxl
import math
pd.set_option('float_format', '{:f}'.format)
import plotly.express as px
import pandas as pd
from datetime import datetime
import plotly.figure_factory as ff

info_df=pd.read_excel(info_path,sheet_name=3)
for i, j in info_df.iterrows(): 
    info_df['Run ID'][i]=j['Run ID'].replace('Run','run')
    #print(j)
#print(info_df['Run ID'])
%store info_df






Dataframes=list()
Dataframes_Jis=list()
Dataframes_Worker=list()
Dataframes_Station=list()
Dataframes_instation=list()
Dataframes_instation_working=list()
Dataframes_time=list()
Runs=list()


directories =os.listdir(path)
for file in directories:
    run=file
    print(run)
    filename=path + '\\' + run + '\\' + run
    Dataframes.append(pd.read_excel(filename+'_analysis_part1.'+ending,sheet_name=0))
    Dataframes_Jis.append(pd.read_excel(filename+'_analysis_part1.'+ending,sheet_name=1))
    Dataframes_Worker.append(pd.read_excel(filename+'_analysis_part2.xlsx',sheet_name=-2))
    Dataframes_Station.append(pd.read_excel(filename+'_analysis_part2.xlsx',sheet_name=-1))
    Dataframes_instation.append(pd.read_excel(filename+'_station_utilization_instation.xlsx'))
    Dataframes_instation_working.append(pd.read_excel(filename+'_station_utilization_working.xlsx'))
    Dataframes_time.append(pd.read_excel(filename+'_analysis_part2.xlsx',sheet_name=0))
    Runs.append(run)
    
%store Dataframes
%store Dataframes_Jis
%store Dataframes_Worker
%store Dataframes_Station
%store Dataframes_instation
%store Dataframes_instation_working
%store Runs

def utilization(df_instation,df_instation_working,df_time):
    global_starttime=int(df_time['arrival timestamp'].iloc[0])
    stations_available=list(df_instation.columns.values)
    #stations_available=['station_2 1', 'station_2 2', 'station_3', 'station_4 1','station_4 2','station_5 1','station_5 2', 'station_6', 'station_7', 'station_1_8']

    dict_list=[]
    for s in range(len(stations_available)):
        history=df_instation[stations_available[s]].values.tolist()
        inplace=False
        for i in range(len(history)):
            if history[i]!=0 and inplace==False:
                start_time=i
                inplace=True
                #written=False

            if history[i]==0 and inplace==True:
                #print(stations_available[s])
                dict_list.append(dict(Task=stations_available[s], Start=datetime.fromtimestamp(1607097208+start_time), Finish=datetime.fromtimestamp(1607097208+i), Resource=('worker'+str(history[i-1]))))
                inplace=False

    for s in range(len(stations_available)):
        history=df_instation_working[stations_available[s]].values.tolist()
        inplace=False
        for i in range(len(history)):
            if history[i]!=0 and inplace==False:
                start_time=i
                inplace=True
                #written=False

            if history[i]==0 and inplace==True:
                #print(stations_available[s])
                dict_list.append(dict(Task=stations_available[s]+'working', Start=datetime.fromtimestamp(1607097208+start_time), Finish=datetime.fromtimestamp(1607097208+i), Resource=('worker'+str(history[i-1]))))
                inplace=False

    #print(dict_list)

    dict_list_worker=[]
    for s in range(len(stations_available)):
        history=df_instation[stations_available[s]].values.tolist()
        inplace=False
        for i in range(len(history)):
            if history[i]!=0 and inplace==False:
                start_time=i
                inplace=True
                #written=False

            if history[i]==0 and inplace==True:
                #print(stations_available[s])
                dict_list_worker.append(dict(Task=('worker'+str(history[i-1])), Start=datetime.fromtimestamp(1607097208+start_time), Finish=datetime.fromtimestamp(1607097208+i), Resource=stations_available[s]))
                inplace=False

    for s in range(len(stations_available)):
        history=df_instation_working[stations_available[s]].values.tolist()
        inplace=False
        for i in range(len(history)):
            if history[i]!=0 and inplace==False:
                start_time=i
                inplace=True
                #written=False

            if history[i]==0 and inplace==True:
                #print(stations_available[s])
                dict_list_worker.append(dict(Task=('worker'+str(history[i-1]))+'working', Start=datetime.fromtimestamp(1607097208+start_time), Finish=datetime.fromtimestamp(1607097208+i), Resource=stations_available[s]))
                inplace=False

    dict_list_worker = sorted(dict_list_worker, key=lambda k: k['Task']) 

    #print(dict_list)

    dict_list = sorted(dict_list, key=lambda k: k['Task']) 
    
    
    return dict_list, dict_list_worker

utilization_list=list()
utilization_list_worker=list()
for i in range(len(Dataframes_instation)):
    dict_list,dict_list_worker=utilization(Dataframes_instation[i],Dataframes_instation_working[i],Dataframes_time[i])
    utilization_list.append(dict_list)
    utilization_list_worker.append(dict_list_worker)
    

%store utilization_list
%store utilization_list_worker
```
